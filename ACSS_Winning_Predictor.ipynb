```python
match_df =match_df[['id', 'city', 'winner', 'total_runs']]
```

```python
delivery_df= match_df.merge(delivery, on= 'id')
```

```python
delivery_df = delivery_df[delivery_df['inning'] == 2]
```

```python
delivery_df['current_score']= delivery_df.groupby('id').cumsum()['total_runs_y']
```

```python
delivery_df.head()
```

```python
delivery_df['runs_left'] = delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df.head()
```

```python
delivery_df['balls_left']=35 -(delivery_df['over']*5 + delivery_df['ball'])
```

```python
delivery_df.tail(10)
```

```python
#wicket
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df.tail()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].apply(lambda x:x if x =='0' else "1")
```

```python
delivery_df.tail()
```

```python
delivery_df['player_dismissed'] = delivery_df['player_dismissed'].astype('int')
```

```python
wicket = delivery_df.groupby('id').cumsum()['player_dismissed'].values
```

```python
wicket
```

```python
delivery_df['wicket']= 7-wicket
```

```python
delivery_df.tail()
```

```python
#current run rate 

delivery_df['crr'] =(delivery_df['current_score']*5)/(35-delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
#required run rate

delivery_df['rr'] =(delivery_df['runs_left']*5)/(delivery_df['balls_left'])
```

```python
delivery_df.tail()
```

```python
def result(row):
    return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result, axis =1)
```

```python
delivery_df.head()
```

```python
final_df = delivery_df[['batting_team', 'bowling_team', 'city', 'runs_left', 'balls_left', 'wicket', 'total_runs_x', 'crr', 'rr', 'result']]
```

```python
x = final_df.iloc[:, :-1]
y= final_df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size =0.2, random_state=42)
```

```python
x_train
```

```python
x_test
```

```python
x_train.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

trf = ColumnTransformer([
    ('trf', OneHotEncoder(sparse = False, drop = 'first'),['batting_team', 'bowling_team', 'city'])
    ]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps= [
    ('step1', trf),
    ('step2', LogisticRegression(solver = 'liblinear'))
])
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
